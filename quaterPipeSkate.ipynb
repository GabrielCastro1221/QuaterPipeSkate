```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

r = 2.4
altura_deseada = 1.8
base_deseada = 2.32

def ecuaciones(theta):
    x = r * np.sin(theta)
    y = r * (1 - np.cos(theta))
    return [x - base_deseada, y - altura_deseada]

theta_sol = fsolve(lambda t: ecuaciones(t)[0]**2 + ecuaciones(t)[1]**2, np.pi / 3)[0]

theta_vals = np.linspace(0, theta_sol, 200)

x = r * np.sin(theta_vals)
y = r * (1 - np.cos(theta_vals))

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Perfil de la rampa (quarter pipe)', color='blue', linewidth=2)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.gca().set_aspect('equal')
plt.xlabel('Distancia horizontal (m)')
plt.ylabel('Altura (m)')
plt.title('Diseo de rampa tipo Quarter Pipe\nRadio = 2.4 m, Altura = 1.8 m, Base = 2.32 m')
plt.legend()
plt.tight_layout()
plt.savefig("quarter_pipe_rampa_parametrizada.png", dpi=300)
plt.show()
```
